```python
from module.MLP import  MLP
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
interval = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(interval, np.sin(interval))
```

```python
# np.random.seed(2)
x = np.random.uniform(-4*np.pi, 4*np.pi, (1000,1))
z = np.sin(x)
neural = MLP(x,z,3,[4,5,6])
errors = neural.train(maxIter = 10000, eta = 0.0001)
plt.plot(errors)
```

```python
x = np.random.uniform(-2*np.pi, 2*np.pi, (1000,1))
predicted = neural.predict(x)

plt.plot(interval, np.sin(interval))
plt.scatter(x,predicted, s= 1, c = "red")
```
