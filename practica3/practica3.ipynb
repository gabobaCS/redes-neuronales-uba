1.1. Regresión y Clasificación:

Utilizando los archivos de datos p13_ej?_regress.csv y p13_ej?_classif.csv provistos para la práctica anterior, encontrar la mejor arquitectura de perceptrón multicapa para estos problemas.

```python
import sys
sys.path.insert(0,"..")
from module.MLP import  MLP
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_ej1 = pd.read_csv("p13_ej1_regress.csv", header=None)


x = df_ej1.iloc[:,:-1].to_numpy()
z = df_ej1.iloc[:,-1].to_numpy().reshape(len(df_ej1),1)

print(type(x), type(z))
print(x.shape, z.shape)
```

```python
neural = MLP(x,z,[3,6,5])
errores = neural.train(minError=1e-3,maxIter=1000, eta=0.0001, batchSize=42)
plt.plot(errores)
plt.show()
```

```python
# print(x[0].shape)
fig, ax = plt.subplots(1, 2, figsize = (15, 15))
fig.set_figheight(6)
ax[0].scatter(x[:,0], x[:,1], c= z.reshape(z.shape[0]), s=10)

test_x = np.random.uniform(-3,3,(5000, 2))
y = neural.predict(test_x)
ax[1].scatter(test_x[:,0], test_x[:,1], c = y.reshape(y.shape[0]), s= 10)
ax[0].set_title("Datos originales")
ax[1].set_title("Datos predichos")
plt.show()
print(neural.error)


```
